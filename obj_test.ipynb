```python
from objloader import objFile

texture_root = 'texture/golf_ball/Golf_Ball_OBJ/'
filename = 'Golf Ball OBJ.obj'

obj = objFile(texture_root, filename, swapyz=True)
print(obj.materials[obj.material])
# print(obj.material)
```

```python
import sys, math
import OpenGL.GL
import OpenGL.GLUT
import OpenGL.GLU


OpenGL.GLUT.glutInit(sys.argv)
for face in obj.faces:
    vertices, normals, texture_coords, material = face
    mat = obj.materials[material]
    for i in range(len(vertices)):
        print(normals[i], texture_coords[i])
        print(vertices[i])
        if normals[i] > 0:
            OpenGL.GL.glNormal3fv(obj.normals[normals[i] - 1])
        if texture_coords[i] > 0:
            OpenGL.GL.glTexCoord2fv(obj.texcoords[texture_coords[i] - 1])
        OpenGL.GL.glVertex3fv(obj.vertices[vertices[i] - 1])
```
